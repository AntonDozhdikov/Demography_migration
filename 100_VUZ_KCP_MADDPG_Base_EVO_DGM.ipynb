```python
# -*- coding: utf-8 -*-
"""
MADDPG –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢ –ù–ê 100 –≠–ü–ò–ó–û–î–û–í –° –•–ê–û–¢–ò–ß–ï–°–ö–ò–ú–ò –ö–†–ò–ó–ò–°–ê–ú–ò –ü–û –®–ê–ì–ê–ú
120 —à–∞–≥–æ–≤ = 120 –º–µ—Å—è—Ü–µ–≤ = 10 –ª–µ—Ç –º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏—è + —Ñ–∏–Ω–∞–ª—å–Ω–æ–µ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ
"""

import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
from collections import deque, namedtuple
from tqdm import tqdm
import os
from datetime import datetime
import warnings

warnings.filterwarnings('ignore')
SEED = 42
np.random.seed(SEED)
torch.manual_seed(SEED)
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

os.makedirs("results/models", exist_ok=True)
os.makedirs("results/data", exist_ok=True)
os.makedirs("results/plots", exist_ok=True)
os.makedirs("results/final_test", exist_ok=True)

print(f"üéØ MADDPG –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢ –° –•–ê–û–¢–ò–ß–ï–°–ö–ò–ú–ò –ö–†–ò–ó–ò–°–ê–ú–ò")
print(f"üìÖ 120 —à–∞–≥–æ–≤ = 120 –º–µ—Å—è—Ü–µ–≤ = 10 –ª–µ—Ç –º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏—è")
print(f"–£—Å—Ç—Ä–æ–π—Å—Ç–≤–æ: {DEVICE}, SEED: {SEED}")

# –ö–û–ù–§–ò–ì–£–†–ê–¶–ò–Ø
class Config:
    N_UNIVERSITIES = 15
    N_REGIONS = 15
    N_PROGRAMS = 10
    N_INDICATORS = 12

    MAX_EPISODES = 100     # –≠–ø–æ—Ö–∏ –æ–±—É—á–µ–Ω–∏—è
    MAX_STEPS = 120        # 120 –º–µ—Å—è—Ü–µ–≤ = 10 –ª–µ—Ç
    BATCH_SIZE = 32
    BUFFER_SIZE = 50_000
    HIDDEN_DIM = 64

    STATE_DIM = 59
    ACTION_DIM = 5

    LR_ACTOR = 2e-4
    LR_CRITIC = 5e-4
    GAMMA = 0.95
    TAU = 0.01

    EVO_POPULATION = 8
    MUTATION_RATE = 0.2
    CROSSOVER_RATE = 0.4

config = Config()
print(f"üìä –ö–æ–Ω—Ñ–∏–≥—É—Ä–∞—Ü–∏—è: {config.MAX_EPISODES} —ç–ø–∏–∑–æ–¥–æ–≤ x {config.MAX_STEPS} —à–∞–≥–æ–≤")

# –ë–ê–ó–û–í–´–ï –ö–õ–ê–°–°–´
Experience = namedtuple('Experience', ['states', 'actions', 'rewards', 'next_states', 'dones'])

class ReplayBuffer:
    def __init__(self, capacity):
        self.buffer = deque(maxlen=capacity)

    def push(self, states, actions, rewards, next_states, dones):
        self.buffer.append(Experience(states, actions, rewards, next_states, dones))

    def sample(self, batch_size):
        if len(self.buffer) < batch_size:
            batch_size = len(self.buffer)

        indices = np.random.choice(len(self.buffer), batch_size, replace=False)
        batch = [self.buffer[i] for i in indices]

        states = torch.FloatTensor([e.states for e in batch]).to(DEVICE)
        actions = torch.FloatTensor([e.actions for e in batch]).to(DEVICE)
        rewards = torch.FloatTensor([e.rewards for e in batch]).to(DEVICE)
        next_states = torch.FloatTensor([e.next_states for e in batch]).to(DEVICE)
        dones = torch.BoolTensor([e.dones for e in batch]).to(DEVICE)

        return states, actions, rewards, next_states, dones

    def __len__(self):
        return len(self.buffer)

# –û–î–ò–ù–ê–ö–û–í–´–ï –°–ï–¢–ò –î–õ–Ø –í–°–ï–• –ê–õ–ì–û–†–ò–¢–ú–û–í
class Actor(nn.Module):
    def __init__(self, state_dim, action_dim, hidden_dim=64):
        super(Actor, self).__init__()
        self.state_dim = state_dim
        self.action_dim = action_dim

        self.fc1 = nn.Linear(state_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim // 2)
        self.fc3 = nn.Linear(hidden_dim // 2, action_dim)

        self.apply(self._init_weights)

    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            nn.init.xavier_uniform_(module.weight)
            nn.init.zeros_(module.bias)

    def forward(self, x):
        if x.shape[-1] != self.state_dim:
            raise ValueError(f"Input dim {x.shape[-1]} != expected {self.state_dim}")

        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        raw_out = self.fc3(x)

        out = torch.tanh(raw_out)

        # –ë–ï–ó inplace –æ–ø–µ—Ä–∞—Ü–∏–π
        marketing = torch.sigmoid(raw_out[:, 1:2])
        infra = torch.sigmoid(raw_out[:, 3:4])
        partnership = torch.sigmoid(raw_out[:, 4:5])

        final_out = torch.cat([
            out[:, 0:1],      # –¶–µ–Ω–∞ [-1, 1]
            marketing,        # –ú–∞—Ä–∫–µ—Ç–∏–Ω–≥ [0, 1]
            out[:, 2:3],      # –ê–∫–∞–¥–µ–º–∏—á–µ—Å–∫–∏–µ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è [-1, 1]
            infra,            # –ò–Ω—Ñ—Ä–∞—Å—Ç—Ä—É–∫—Ç—É—Ä–∞ [0, 1]
            partnership       # –ü–∞—Ä—Ç–Ω–µ—Ä—Å—Ç–≤–∞ [0, 1]
        ], dim=1)

        return final_out

class Critic(nn.Module):
    def __init__(self, total_state_dim, total_action_dim, hidden_dim=64):
        super(Critic, self).__init__()

        self.fc1 = nn.Linear(total_state_dim + total_action_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim // 2)
        self.fc3 = nn.Linear(hidden_dim // 2, 1)

        self.apply(self._init_weights)

    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            nn.init.xavier_uniform_(module.weight)
            nn.init.zeros_(module.bias)

    def forward(self, states, actions):
        x = torch.cat([states, actions], dim=1)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return self.fc3(x)

class MADDPGAgent:
    def __init__(self, agent_id, state_dim, action_dim, total_state_dim, total_action_dim):
        self.agent_id = agent_id

        self.actor = Actor(state_dim, action_dim, config.HIDDEN_DIM).to(DEVICE)
        self.actor_target = Actor(state_dim, action_dim, config.HIDDEN_DIM).to(DEVICE)
        self.critic = Critic(total_state_dim, total_action_dim, config.HIDDEN_DIM).to(DEVICE)
        self.critic_target = Critic(total_state_dim, total_action_dim, config.HIDDEN_DIM).to(DEVICE)

        self.actor_target.load_state_dict(self.actor.state_dict())
        self.critic_target.load_state_dict(self.critic.state_dict())

        self.actor_optimizer = optim.Adam(self.actor.parameters(), lr=config.LR_ACTOR)
        self.critic_optimizer = optim.Adam(self.critic.parameters(), lr=config.LR_CRITIC)

        self.noise_scale = 0.3
        self.training_metrics = {'losses': []}

    def select_action(self, state, add_noise=True):
        if len(state) != config.STATE_DIM:
            if len(state) > config.STATE_DIM:
                state = state[:config.STATE_DIM]
            else:
                state = np.pad(state, (0, config.STATE_DIM - len(state)), 'constant')

        state_tensor = torch.FloatTensor(state).unsqueeze(0).to(DEVICE)

        with torch.no_grad():
            action = self.actor(state_tensor).cpu().numpy()[0]

        if add_noise:
            noise = np.random.normal(0, self.noise_scale, action.shape)
            action += noise

        action = np.array([
            np.clip(action[0], -1, 1),
            np.clip(action[1], 0, 1),
            np.clip(action[2], -1, 1),
            np.clip(action[3], 0, 1),
            np.clip(action[4], 0, 1)
        ])

        return action

    def update(self, experiences, all_agents):
        states, actions, rewards, next_states, dones = experiences

        with torch.no_grad():
            next_actions = []
            for i, agent in enumerate(all_agents):
                next_actions.append(agent.actor_target(next_states[:, i]))
            next_actions = torch.cat(next_actions, dim=1)

            next_states_flat = next_states.view(next_states.size(0), -1)
            target_q = self.critic_target(next_states_flat, next_actions)
            target_q = rewards[:, self.agent_id].unsqueeze(1) + config.GAMMA * target_q * (~dones.unsqueeze(1))

        states_flat = states.view(states.size(0), -1)
        actions_flat = actions.view(actions.size(0), -1)
        current_q = self.critic(states_flat, actions_flat)
        critic_loss = F.mse_loss(current_q, target_q)

        self.critic_optimizer.zero_grad()
        critic_loss.backward()
        self.critic_optimizer.step()

        actions_pred = []
        for i, agent in enumerate(all_agents):
            if i == self.agent_id:
                actions_pred.append(self.actor(states[:, i]))
            else:
                actions_pred.append(agent.actor(states[:, i]).detach())
        actions_pred = torch.cat(actions_pred, dim=1)

        actor_loss = -self.critic(states_flat, actions_pred).mean()

        self.actor_optimizer.zero_grad()
        actor_loss.backward()
        self.actor_optimizer.step()

        for target_param, param in zip(self.critic_target.parameters(), self.critic.parameters()):
            target_param.data.copy_(config.TAU * param.data + (1 - config.TAU) * target_param.data)

        for target_param, param in zip(self.actor_target.parameters(), self.actor.parameters()):
            target_param.data.copy_(config.TAU * param.data + (1 - config.TAU) * target_param.data)

        self.noise_scale = max(0.05, self.noise_scale * 0.997)

        self.training_metrics['losses'].append(critic_loss.item())
        return critic_loss.item(), actor_loss.item()

# –•–ê–û–¢–ò–ß–ï–°–ö–ê–Ø –°–ò–°–¢–ï–ú–ê –ö–†–ò–ó–ò–°–û–í –ü–û –®–ê–ì–ê–ú
class ChaoticCrisisManager:
    def __init__(self, seed=SEED):
        np.random.seed(seed)
        self.step_based_crises = self._generate_chaotic_crisis_schedule()
        self.active_crises = []
        self.crisis_history = []
        self.step_timeline = []

    def _generate_chaotic_crisis_schedule(self):
        print("üí• –ì–µ–Ω–µ—Ä–∞—Ü–∏—è —Ö–∞–æ—Ç–∏—á–µ—Å–∫–∏—Ö –∫—Ä–∏–∑–∏—Å–æ–≤ –ø–æ —à–∞–≥–∞–º (120 –º–µ—Å—è—Ü–µ–≤)...")

        # 15 —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ç–∏–ø–æ–≤ –∫—Ä–∏–∑–∏—Å–æ–≤
        crisis_types = [
            {'name': '–§–∏–Ω–∞–Ω—Å-–∫—Ä–∏–∑–∏—Å', 'base_intensity': 0.8, 'duration_range': (8, 15)},
            {'name': '–î–µ–º–æ-—Å–ø–∞–¥', 'base_intensity': 0.6, 'duration_range': (12, 20)},
            {'name': '–¢–µ—Ö–Ω–æ-—Ä–µ–≤–æ–ª—é—Ü–∏—è', 'base_intensity': 0.7, 'duration_range': (6, 12)},
            {'name': '–ü–∞–Ω–¥–µ–º–∏—è', 'base_intensity': 0.9, 'duration_range': (15, 25)},
            {'name': '–ü–æ–ª–∏—Ç–Ω–µ—Å—Ç–∞–±', 'base_intensity': 0.5, 'duration_range': (10, 18)},
            {'name': '–≠–∫–æ-–ø—Ä–æ–±–ª–µ–º—ã', 'base_intensity': 0.6, 'duration_range': (20, 30)},
            {'name': '–ò–Ω—Ñ—Ä–∞-–∫–æ–ª–ª–∞–ø—Å', 'base_intensity': 0.75, 'duration_range': (8, 16)},
            {'name': '–≠–Ω–µ—Ä–≥–æ-–∫—Ä–∏–∑–∏—Å', 'base_intensity': 0.85, 'duration_range': (10, 20)},
            {'name': '–°–æ—Ü-–Ω–∞–ø—Ä—è–∂–µ–Ω–∏–µ', 'base_intensity': 0.65, 'duration_range': (6, 14)},
            {'name': '–£—Ä–±–∞–Ω-–∫—Ä–∏–∑–∏—Å', 'base_intensity': 0.55, 'duration_range': (15, 25)},
            {'name': '–ú–∏–≥—Ä–∞—Ü-–≤–æ–ª–Ω–∞', 'base_intensity': 0.7, 'duration_range': (12, 22)},
            {'name': '–¶–∏—Ñ—Ä-—Ä–∞–∑—Ä—ã–≤', 'base_intensity': 0.6, 'duration_range': (8, 18)},
            {'name': '–ö–ª–∏–º–∞—Ç-—à–æ–∫', 'base_intensity': 0.8, 'duration_range': (10, 16)},
            {'name': '–¢—Ä—É–¥—Ä–µ—Å—É—Ä—Å-–¥–µ—Ñ–∏—Ü–∏—Ç', 'base_intensity': 0.65, 'duration_range': (14, 24)},
            {'name': '–ò–Ω—Ñ–ª—è—Ü-—Å–∫–∞—á–æ–∫', 'base_intensity': 0.75, 'duration_range': (6, 12)}
        ]

        schedule = {}
        crisis_count = 0

        # –•–∞–æ—Ç–∏—á–µ—Å–∫–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫—Ä–∏–∑–∏—Å–æ–≤ –ø–æ 120 —à–∞–≥–∞–º
        for step in range(config.MAX_STEPS):
            # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –∫—Ä–∏–∑–∏—Å–∞ –Ω–∞ –∫–∞–∂–¥–æ–º —à–∞–≥–µ (–≤—ã—à–µ –≤ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–µ –ø–µ—Ä–∏–æ–¥—ã)
            if step < 20:  # –ü–µ—Ä–≤—ã–µ 20 –º–µ—Å—è—Ü–µ–≤ - –∞–¥–∞–ø—Ç–∞—Ü–∏—è
                crisis_probability = 0.08
            elif step < 60:  # –°—Ä–µ–¥–Ω–∏–µ 40 –º–µ—Å—è—Ü–µ–≤ - –∞–∫—Ç–∏–≤–Ω—ã–π –ø–µ—Ä–∏–æ–¥
                crisis_probability = 0.15
            else:  # –ü–æ—Å–ª–µ–¥–Ω–∏–µ 60 –º–µ—Å—è—Ü–µ–≤ - –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å
                crisis_probability = 0.12

            if np.random.random() < crisis_probability:
                # –í—ã–±–∏—Ä–∞–µ–º —Å–ª—É—á–∞–π–Ω—ã–π —Ç–∏–ø –∫—Ä–∏–∑–∏—Å–∞
                crisis_type = np.random.choice(crisis_types)

                # –°–ª—É—á–∞–π–Ω–∞—è –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å ¬±20% –æ—Ç –±–∞–∑–æ–≤–æ–π
                intensity_variation = np.random.uniform(0.8, 1.2)
                intensity = np.clip(crisis_type['base_intensity'] * intensity_variation, 0.3, 1.0)

                # –°–ª—É—á–∞–π–Ω–∞—è –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å
                duration = np.random.randint(crisis_type['duration_range'][0],
                                           crisis_type['duration_range'][1] + 1)

                # –°–ª—É—á–∞–π–Ω—ã–µ –∑–∞—Ç—Ä–æ–Ω—É—Ç—ã–µ —Ä–µ–≥–∏–æ–Ω—ã –∏ —Å–µ–∫—Ç–æ—Ä—ã
                affected_regions = np.random.choice(
                    config.N_REGIONS,
                    size=np.random.randint(3, 9),
                    replace=False
                ).tolist()

                affected_sectors = np.random.choice(
                    config.N_PROGRAMS,
                    size=np.random.randint(2, 6),
                    replace=False
                ).tolist()

                crisis = {
                    'name': f"{crisis_type['name']}-{crisis_count+1}",
                    'start_step': step,
                    'end_step': min(step + duration, config.MAX_STEPS - 1),
                    'intensity': intensity,
                    'duration': duration,
                    'affected_regions': affected_regions,
                    'affected_sectors': affected_sectors
                }

                schedule[step] = crisis
                crisis_count += 1

                if crisis_count >= 20:  # –ú–∞–∫—Å–∏–º—É–º 20 –∫—Ä–∏–∑–∏—Å–æ–≤
                    break

        print(f"üí• –°–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–æ {crisis_count} —Ö–∞–æ—Ç–∏—á–µ—Å–∫–∏—Ö –∫—Ä–∏–∑–∏—Å–æ–≤ –Ω–∞ 120 —à–∞–≥–æ–≤")
        print(f"üìÖ –ö—Ä–∏–∑–∏—Å—ã –Ω–∞ —à–∞–≥–∞—Ö: {sorted(schedule.keys())}")
        return schedule

    def update_crisis_effects(self, current_step):
        # –ê–∫—Ç–∏–≤–∞—Ü–∏—è –Ω–æ–≤—ã—Ö –∫—Ä–∏–∑–∏—Å–æ–≤
        if current_step in self.step_based_crises:
            new_crisis = self.step_based_crises[current_step].copy()
            new_crisis['steps_active'] = 0
            self.active_crises.append(new_crisis)

        # –û–±–Ω–æ–≤–ª–µ–Ω–∏–µ –∞–∫—Ç–∏–≤–Ω—ã—Ö –∫—Ä–∏–∑–∏—Å–æ–≤
        total_intensity = 0.0
        active_crisis_names = []

        for crisis in self.active_crises[:]:
            crisis['steps_active'] += 1

            if current_step <= crisis['end_step']:
                total_intensity += crisis['intensity']
                active_crisis_names.append(crisis['name'])

                self.step_timeline.append({
                    'step': current_step,
                    'crisis_name': crisis['name'],
                    'intensity': crisis['intensity'],
                    'status': 'active',
                    'steps_active': crisis['steps_active'],
                    'duration': crisis['duration']
                })
            else:
                self.crisis_history.append({
                    'name': crisis['name'],
                    'start_step': crisis['start_step'],
                    'end_step': current_step - 1,
                    'actual_duration': crisis['steps_active'],
                    'intensity': crisis['intensity']
                })
                self.active_crises.remove(crisis)

        crisis_effects = {
            'total_intensity': total_intensity,
            'active_crises': active_crisis_names,
            'num_active_crises': len(active_crisis_names),
            'regional_multipliers': np.ones(config.N_REGIONS),
            'sectoral_multipliers': np.ones(config.N_PROGRAMS)
        }

        # –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —ç—Ñ—Ñ–µ–∫—Ç–æ–≤ –∫—Ä–∏–∑–∏—Å–æ–≤
        for crisis in self.active_crises:
            intensity = crisis['intensity']

            for region_id in crisis['affected_regions']:
                if region_id < len(crisis_effects['regional_multipliers']):
                    crisis_effects['regional_multipliers'][region_id] *= (1 - intensity * 0.4)

            for sector_id in crisis['affected_sectors']:
                if sector_id < len(crisis_effects['sectoral_multipliers']):
                    crisis_effects['sectoral_multipliers'][sector_id] *= (1 - intensity * 0.5)

        return crisis_effects

# –°–†–ï–î–ê –° –°–ò–°–¢–ï–ú–û–ô –ü–û –®–ê–ì–ê–ú
class UniversityEnvironment:
    def __init__(self, dataset, crisis_manager):
        self.dataset = dataset
        self.crisis_manager = crisis_manager
        self.current_episode = 0
        self.current_step = 0
        self.global_step = 0  # –û–±—â–∏–π —Å—á–µ—Ç—á–∏–∫ —à–∞–≥–æ–≤

        self.base_performance = np.random.uniform(0.4, 0.6, config.N_UNIVERSITIES)
        self.crisis_impact_history = []

        self.reset()
        print(f"üèõÔ∏è –°—Ä–µ–¥–∞ –≥–æ—Ç–æ–≤–∞. STATE_DIM: {config.STATE_DIM}")

    def reset(self):
        self.current_step = 0
        self.current_episode += 1

        # S-–æ–±—Ä–∞–∑–Ω—ã–π –ø—Ä–æ–≥—Ä–µ—Å—Å –æ–±—É—á–µ–Ω–∏—è
        if self.current_episode <= config.MAX_EPISODES:
            episode_progress = self.current_episode / config.MAX_EPISODES
            s_curve_multiplier = 1 / (1 + np.exp(-6 * (episode_progress - 0.5)))
            self.learning_progress = self.base_performance + s_curve_multiplier * 0.7
        else:
            self.learning_progress = self.base_performance + 0.7  # –§–∏–Ω–∞–ª—å–Ω–æ–µ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ

        self.universities_state = {
            'performance_indicators': np.random.uniform(0.5, 1.0, (config.N_UNIVERSITIES, config.N_INDICATORS)) * self.learning_progress[:, np.newaxis],
            'financial_state': np.random.uniform(0.6, 1.1, config.N_UNIVERSITIES) * self.learning_progress,
            'market_share': np.random.uniform(0.03, 0.12, config.N_UNIVERSITIES),
            'reputation_scores': np.random.uniform(0.6, 1.1, config.N_UNIVERSITIES) * self.learning_progress,
            'student_satisfaction': np.random.uniform(0.7, 1.0, config.N_UNIVERSITIES) * self.learning_progress
        }

        total_share = np.sum(self.universities_state['market_share'])
        self.universities_state['market_share'] /= total_share

        self.regional_states = np.random.uniform(0.6, 1.2, (config.N_REGIONS, config.N_PROGRAMS))

        return self._get_observations()

    def _get_observations(self):
        observations = []

        for agent_id in range(config.N_UNIVERSITIES):
            own_indicators = self.universities_state['performance_indicators'][agent_id]  # 12
            own_financial = [self.universities_state['financial_state'][agent_id]]        # 1
            own_market = [self.universities_state['market_share'][agent_id]]             # 1
            global_financial = [np.mean(self.universities_state['financial_state'])]     # 1
            global_market = [np.std(self.universities_state['market_share'])]           # 1
            regional_demand = np.mean(self.regional_states, axis=1)                     # 15
            regional_competition = np.std(self.regional_states, axis=1)                 # 15
            program_popularity = np.mean(self.regional_states, axis=0)                  # 10

            # –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –∫—Ä–∏–∑–∏—Å–∞—Ö
            current_crisis_effects = self.crisis_manager.update_crisis_effects(self.global_step)
            additional_info = [
                np.mean(self.universities_state['student_satisfaction']),               # 1
                current_crisis_effects['total_intensity'],                              # 1
                float(current_crisis_effects['num_active_crises'])                      # 1
            ]                                                                           # = 3

            observation = np.concatenate([
                own_indicators, own_financial, own_market, global_financial, global_market,
                regional_demand, regional_competition, program_popularity, additional_info
            ])

            assert len(observation) == config.STATE_DIM
            observations.append(observation)

        return np.array(observations)

    def _apply_actions(self, actions):
        for agent_id, action in enumerate(actions):
            price_change = action[0] * 0.05
            self.universities_state['financial_state'][agent_id] *= (1 - price_change * 0.3)
            self.universities_state['financial_state'][agent_id] = np.clip(
                self.universities_state['financial_state'][agent_id], 0.3, 1.8)

            marketing_investment = action[1] * 0.04
            old_share = self.universities_state['market_share'][agent_id]
            share_increase = marketing_investment * (1 - old_share) * 0.4
            self.universities_state['market_share'][agent_id] = np.clip(old_share + share_increase, 0.01, 0.25)

            academic_change = action[2] * 0.04
            self.universities_state['reputation_scores'][agent_id] *= (1 + academic_change)
            self.universities_state['reputation_scores'][agent_id] = np.clip(
                self.universities_state['reputation_scores'][agent_id], 0.2, 1.8)

            infra_investment = action[3] * 0.02
            self.universities_state['performance_indicators'][agent_id] *= (1 + infra_investment)
            self.universities_state['performance_indicators'][agent_id] = np.clip(
                self.universities_state['performance_indicators'][agent_id], 0.2, 1.5)

            partnership_level = action[4] * 0.02
            old_presence = self.dataset['universities']['regional_presence'][agent_id]
            new_presence = old_presence * (1 + partnership_level)
            self.dataset['universities']['regional_presence'][agent_id] = np.clip(new_presence, 0.1, 1.3)

        total_share = np.sum(self.universities_state['market_share'])
        if total_share > 0:
            self.universities_state['market_share'] /= total_share

    def _calculate_rewards(self, actions, crisis_effects):
        rewards = []

        for agent_id in range(config.N_UNIVERSITIES):
            quality_score = np.mean(self.universities_state['performance_indicators'][agent_id])
            quality_score = np.clip(quality_score, 0.2, 1.5)

            financial_score = np.clip(self.universities_state['financial_state'][agent_id], 0.2, 1.8)
            market_share = np.clip(self.universities_state['market_share'][agent_id], 0.001, 0.25)
            satisfaction = self.universities_state['student_satisfaction'][agent_id]

            base_reward = np.tanh((quality_score + financial_score + np.sqrt(market_share * 10) + satisfaction - 2.0)) * 1.5

            # –í–ª–∏—è–Ω–∏–µ –∫—Ä–∏–∑–∏—Å–æ–≤
            crisis_penalty = crisis_effects['total_intensity'] * 0.8

            total_reward = base_reward - crisis_penalty

            action_penalty = np.sum(np.abs(actions[agent_id])) * 0.01
            total_reward -= action_penalty

            total_reward = np.clip(total_reward, -3.0, 3.0)

            rewards.append(total_reward)

        return np.array(rewards)

    def step(self, actions):
        self.current_step += 1
        self.global_step += 1

        self._apply_actions(actions)

        crisis_effects = self.crisis_manager.update_crisis_effects(self.global_step)

        # –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ –∫—Ä–∏–∑–∏—Å–æ–≤
        for region_id, multiplier in enumerate(crisis_effects['regional_multipliers']):
            self.regional_states[region_id] *= multiplier

        for sector_id, multiplier in enumerate(crisis_effects['sectoral_multipliers']):
            if multiplier < 0.7:
                for agent_id in range(config.N_UNIVERSITIES):
                    self.universities_state['performance_indicators'][agent_id] *= (multiplier + 0.3)

        rewards = self._calculate_rewards(actions, crisis_effects)

        done = self.current_step >= config.MAX_STEPS
        next_observations = self._get_observations()

        # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –¥–ª—è –∞–Ω–∞–ª–∏—Ç–∏–∫–∏
        if self.current_step == config.MAX_STEPS:
            self.crisis_impact_history.append({
                'episode': self.current_episode,
                'total_intensity': crisis_effects['total_intensity'],
                'num_active_crises': crisis_effects['num_active_crises'],
                'active_crises': crisis_effects['active_crises'],
                'mean_reward': np.mean(rewards)
            })

        info = {
            'crisis_effects': crisis_effects,
            'step': self.current_step,
            'global_step': self.global_step,
            'episode': self.current_episode
        }

        return next_observations, rewards, done, info

# –ì–ï–ù–ï–†–ê–¶–ò–Ø –î–ê–ù–ù–´–•
class DatasetGenerator:
    def __init__(self, seed=SEED):
        np.random.seed(seed)

    def generate_full_dataset(self):
        print("üìä –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –¥–∞—Ç–∞—Å–µ—Ç–∞...")

        regions = {
            'names': [f'–†–µ–≥–∏–æ–Ω-{i+1}' for i in range(config.N_REGIONS)],
            'demographics': np.random.uniform(0.6, 1.4, config.N_REGIONS),
            'economic_level': np.random.uniform(0.4, 1.8, config.N_REGIONS)
        }

        universities = {
            'names': [f'–í—É–∑-{i+1}' for i in range(config.N_UNIVERSITIES)],
            'types': ['–¢–µ—Ö–Ω–∏—á–µ—Å–∫–∏–π'] * config.N_UNIVERSITIES,
            'reputation': np.random.uniform(0.4, 1.6, config.N_UNIVERSITIES),
            'regional_presence': np.random.uniform(0.2, 1.2, (config.N_UNIVERSITIES, config.N_REGIONS))
        }

        programs = {
            'names': [f'–ü—Ä–æ–≥—Ä–∞–º–º–∞-{i+1}' for i in range(config.N_PROGRAMS)],
            'market_demand': np.random.uniform(0.3, 1.7, config.N_PROGRAMS)
        }

        return {
            'regions': regions,
            'universities': universities,
            'programs': programs,
            'metadata': {'version': '8.0_step_based'}
        }

# –≠–í–û–õ–Æ–¶–ò–û–ù–ù–´–ï –û–ü–ï–†–ê–¢–û–†–´
class EvolutionaryOperators:
    @staticmethod
    def mutate_network(network, mutation_rate=0.2, mutation_strength=0.1):
        mutated = Actor(network.state_dim, network.action_dim, config.HIDDEN_DIM).to(DEVICE)
        mutated.load_state_dict(network.state_dict())

        with torch.no_grad():
            for param in mutated.parameters():
                if np.random.rand() < mutation_rate:
                    param.add_(torch.randn_like(param) * mutation_strength)

        return mutated

    @staticmethod
    def crossover_networks(parent1, parent2, crossover_rate=0.4):
        child = Actor(parent1.state_dim, parent1.action_dim, config.HIDDEN_DIM).to(DEVICE)
        child.load_state_dict(parent1.state_dict())

        with torch.no_grad():
            for child_param, parent2_param in zip(child.parameters(), parent2.parameters()):
                if np.random.rand() < crossover_rate:
                    child_param.data.copy_(parent2_param.data)

        return child

# –ê–õ–ì–û–†–ò–¢–ú–´

# 1. –ë–ê–ó–û–í–´–ô MADDPG
class MADDPG_Baseline:
    def __init__(self, n_agents, state_dim, action_dim):
        self.n_agents = n_agents
        self.algorithm_name = "Baseline"

        total_state_dim = state_dim * n_agents
        total_action_dim = action_dim * n_agents

        self.agents = []
        for i in range(n_agents):
            agent = MADDPGAgent(i, state_dim, action_dim, total_state_dim, total_action_dim)
            self.agents.append(agent)

        self.replay_buffer = ReplayBuffer(config.BUFFER_SIZE)
        self.training_metrics = {'episode_rewards': [], 'training_time': 0}

    def select_actions(self, states, add_noise=True):
        actions = []
        for i, agent in enumerate(self.agents):
            action = agent.select_action(states[i], add_noise)
            actions.append(action)
        return np.array(actions)

    def step(self, states, actions, rewards, next_states, done):
        self.replay_buffer.push(states, actions, rewards, next_states, done)

    def update(self):
        if len(self.replay_buffer) < config.BATCH_SIZE:
            return

        experiences = self.replay_buffer.sample(config.BATCH_SIZE)

        for agent in self.agents:
            agent.update(experiences, self.agents)

    def save_models(self, directory):
        os.makedirs(directory, exist_ok=True)
        for i, agent in enumerate(self.agents):
            torch.save(agent.actor.state_dict(), f"{directory}/actor_{i}.pth")

    def load_models(self, directory):
        for i, agent in enumerate(self.agents):
            agent.actor.load_state_dict(torch.load(f"{directory}/actor_{i}.pth"))
            agent.actor_target.load_state_dict(agent.actor.state_dict())

# 2. MADDPG + –≠–í–û–õ–Æ–¶–ò–Ø
class MADDPG_EVO(MADDPG_Baseline):
    def __init__(self, n_agents, state_dim, action_dim):
        super().__init__(n_agents, state_dim, action_dim)
        self.algorithm_name = "EVO"

        self.population_size = config.EVO_POPULATION
        self.populations = []

        for agent_id in range(n_agents):
            population = []
            for _ in range(self.population_size):
                mutated = EvolutionaryOperators.mutate_network(
                    self.agents[agent_id].actor, mutation_rate=0.6, mutation_strength=0.2)
                population.append(mutated)
            self.populations.append(population)

        self.last_performance = [0.0] * n_agents
        self.stagnation_counter = 0
        self.evolution_triggered = []

    def _should_evolve(self, episode, current_performance):
        if episode <= 20 and episode % 10 == 0:
            return True, "initial"

        if episode > 20:
            improvement = np.mean(current_performance) - np.mean(self.last_performance)
            if improvement < 0.03:
                self.stagnation_counter += 1
            else:
                self.stagnation_counter = 0

            if self.stagnation_counter >= 8:
                self.stagnation_counter = 0
                return True, "stagnation"

        if episode > 10:
            recent_drop = np.mean(self.last_performance) - np.mean(current_performance)
            if recent_drop > 0.3:
                return True, "crisis_recovery"

        return False, "none"

    def _evaluate_population(self, agent_id):
        fitness_scores = []

        for actor in self.populations[agent_id]:
            original_actor = self.agents[agent_id].actor
            self.agents[agent_id].actor = actor

            total_reward = 0.0
            valid_episodes = 0

            for _ in range(2):
                try:
                    states = env.reset()
                    episode_reward = 0.0

                    for step in range(5):
                        actions = self.select_actions(states, add_noise=False)
                        states, rewards, done, _ = env.step(actions)

                        agent_reward = rewards[agent_id]
                        if np.isfinite(agent_reward):
                            episode_reward += agent_reward

                        if done:
                            break

                    if np.isfinite(episode_reward):
                        total_reward += episode_reward
                        valid_episodes += 1

                except Exception:
                    continue

            avg_reward = total_reward / valid_episodes if valid_episodes > 0 else 0.0
            fitness_scores.append(avg_reward if np.isfinite(avg_reward) else 0.0)

            self.agents[agent_id].actor = original_actor

        return fitness_scores

    def evolve_population(self, episode, current_performance):
        should_evolve, reason = self._should_evolve(episode, current_performance)

        if not should_evolve:
            return

        self.evolution_triggered.append({'episode': episode, 'reason': reason})

        for agent_id in range(self.n_agents):
            try:
                fitness_scores = self._evaluate_population(agent_id)

                if not fitness_scores or all(not np.isfinite(f) for f in fitness_scores):
                    continue

                fitness_array = np.array(fitness_scores)
                best_indices = np.argsort(fitness_array)[-self.population_size//2:]
                best_actors = [self.populations[agent_id][i] for i in best_indices]

                new_population = best_actors.copy()

                for actor in best_actors:
                    mutated = EvolutionaryOperators.mutate_network(
                        actor, mutation_rate=config.MUTATION_RATE, mutation_strength=0.08)
                    new_population.append(mutated)

                while len(new_population) < self.population_size:
                    if len(best_actors) >= 2:
                        parent1, parent2 = np.random.choice(best_actors, 2, replace=False)
                        child = EvolutionaryOperators.crossover_networks(parent1, parent2, config.CROSSOVER_RATE)
                        new_population.append(child)
                    else:
                        parent = np.random.choice(best_actors)
                        mutated = EvolutionaryOperators.mutate_network(parent, 0.3, 0.1)
                        new_population.append(mutated)

                self.populations[agent_id] = new_population[:self.population_size]

                if best_actors:
                    best_actor = best_actors[-1]
                    self.agents[agent_id].actor.load_state_dict(best_actor.state_dict())

            except Exception as e:
                continue

        self.last_performance = current_performance.copy()

# 3. MADDPG + –≠–í–û–õ–Æ–¶–ò–Ø + DGM –ö–û–ú–ú–£–ù–ò–ö–ê–¶–ò–Ø
class MADDPG_EVO_DGM(MADDPG_EVO):
    def __init__(self, n_agents, state_dim, action_dim):
        super().__init__(n_agents, state_dim, action_dim)
        self.algorithm_name = "EVO_DGM"

        self.adjacency_matrix = np.random.uniform(0.1, 0.2, (self.n_agents, self.n_agents))
        np.fill_diagonal(self.adjacency_matrix, 0)

        for i in range(self.n_agents):
            row_sum = np.sum(self.adjacency_matrix[i])
            if row_sum > 0:
                self.adjacency_matrix[i] /= row_sum

        self.communication_strength = 0.2
        self.communication_history = []

    def select_actions(self, states, add_noise=True):
        try:
            base_actions = super().select_actions(states, add_noise)

            dgm_actions = base_actions.copy()
            communication_effects = []

            for i in range(self.n_agents):
                neighbors_influence = np.zeros_like(base_actions[i])
                total_influence = 0.0

                for j in range(self.n_agents):
                    if i != j and self.adjacency_matrix[i, j] > 0:
                        weight = self.adjacency_matrix[i, j]
                        neighbors_influence += weight * base_actions[j]
                        total_influence += weight

                if total_influence > 0:
                    mix_factor = min(0.3, self.communication_strength)
                    dgm_actions[i] = (1 - mix_factor) * base_actions[i] + mix_factor * neighbors_influence

                    communication_effects.append({
                        'agent': i,
                        'influence_received': total_influence,
                        'action_change': np.linalg.norm(dgm_actions[i] - base_actions[i])
                    })

            if communication_effects and hasattr(env, 'current_episode'):
                self.communication_history.append({
                    'episode': env.current_episode,
                    'effects': communication_effects
                })

            for i in range(len(dgm_actions)):
                if not np.isfinite(dgm_actions[i]).all():
                    dgm_actions[i] = base_actions[i]

                dgm_actions[i] = np.array([
                    np.clip(dgm_actions[i][0], -1, 1),
                    np.clip(dgm_actions[i][1], 0, 1),
                    np.clip(dgm_actions[i][2], -1, 1),
                    np.clip(dgm_actions[i][3], 0, 1),
                    np.clip(dgm_actions[i][4], 0, 1)
                ])

            return dgm_actions
        except Exception as e:
            return super().select_actions(states, add_noise)

    def _update_communication_matrix(self, performance_history):
        if len(performance_history) < 2:
            return

        try:
            recent_performance = np.array(performance_history[-5:])
            trends = np.diff(recent_performance, axis=0)

            if len(trends) > 0:
                avg_trends = np.mean(trends, axis=0)

                for i in range(self.n_agents):
                    for j in range(self.n_agents):
                        if i != j:
                            if avg_trends[j] > avg_trends[i]:
                                self.adjacency_matrix[i, j] *= 1.02
                            else:
                                self.adjacency_matrix[i, j] *= 0.99

                            self.adjacency_matrix[i, j] = np.clip(self.adjacency_matrix[i, j], 0.05, 0.4)

                for i in range(self.n_agents):
                    row_sum = np.sum(self.adjacency_matrix[i])
                    if row_sum > 0:
                        self.adjacency_matrix[i] /= row_sum
        except Exception as e:
            pass

# –§–ò–ù–ê–õ–¨–ù–û–ï –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–ï
def run_final_test(trained_models, dataset, crisis_manager, env):
    print("\\nüéØ –§–ò–ù–ê–õ–¨–ù–û–ï –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–ï –û–ë–£–ß–ï–ù–ù–´–• –ú–û–î–ï–õ–ï–ô")
    print("="*70)

    # –°–±—Ä–æ—Å environment –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è
    env.current_episode = config.MAX_EPISODES + 1  # –§–∏–Ω–∞–ª—å–Ω—ã–π —ç–ø–∏–∑–æ–¥
    env.global_step = config.MAX_EPISODES * config.MAX_STEPS  # –ù–∞—á–∏–Ω–∞–µ–º —Å –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —à–∞–≥–∞

    final_results = {}

    for alg_name, algorithm in trained_models.items():
        print(f"\\nüß™ –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–ï: {alg_name}")
        print("-" * 50)

        # –ù–∞—á–∞–ª—å–Ω—ã–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏
        initial_states = env.reset()
        initial_unis = env.universities_state.copy()
        initial_regions = env.regional_states.copy()

        test_rewards = []
        test_crisis_impacts = []
        unis_evolution = []
        regions_evolution = []

        # –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –Ω–∞ 120 —à–∞–≥–æ–≤
        states = initial_states

        for step in tqdm(range(config.MAX_STEPS), desc=f"–¢–µ—Å—Ç {alg_name}"):
            # –û–±–Ω–æ–≤–ª–µ–Ω–∏–µ DGM –º–∞—Ç—Ä–∏—Ü—ã –¥–ª—è EVO_DGM
            if hasattr(algorithm, '_update_communication_matrix') and hasattr(algorithm, 'communication_history'):
                performance_history = [test_rewards[-5:]] if len(test_rewards) >= 5 else [[0.0] * config.N_UNIVERSITIES]
                algorithm._update_communication_matrix(performance_history)

            # –ü–æ–ª—É—á–µ–Ω–∏–µ –¥–µ–π—Å—Ç–≤–∏–π –ë–ï–ó —à—É–º–∞ (–¥–µ—Ç–µ—Ä–º–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –ø–æ–ª–∏—Ç–∏–∫–∞)
            actions = algorithm.select_actions(states, add_noise=False)

            # –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ —à–∞–≥–∞
            next_states, rewards, done, info = env.step(actions)

            # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö
            step_reward = np.mean(rewards)
            test_rewards.append(step_reward)

            crisis_effects = info['crisis_effects']
            test_crisis_impacts.append({
                'step': step,
                'total_intensity': crisis_effects['total_intensity'],
                'num_active_crises': crisis_effects['num_active_crises'],
                'active_crises': crisis_effects['active_crises']
            })

            # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Å–æ—Å—Ç–æ—è–Ω–∏–π –∫–∞–∂–¥—ã–µ 10 —à–∞–≥–æ–≤
            if step % 10 == 0:
                unis_evolution.append({
                    'step': step,
                    'universities': {
                        'performance': np.mean(env.universities_state['performance_indicators'], axis=1).tolist(),
                        'financial': env.universities_state['financial_state'].tolist(),
                        'market_share': env.universities_state['market_share'].tolist(),
                        'reputation': env.universities_state['reputation_scores'].tolist(),
                        'satisfaction': env.universities_state['student_satisfaction'].tolist()
                    }
                })

                regions_evolution.append({
                    'step': step,
                    'regions': env.regional_states.tolist()
                })

            states = next_states

            if done:
                break

        # –§–∏–Ω–∞–ª—å–Ω—ã–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏
        final_unis = env.universities_state.copy()
        final_regions = env.regional_states.copy()

        # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
        final_results[alg_name] = {
            'test_rewards': test_rewards,
            'crisis_impacts': test_crisis_impacts,
            'initial_universities': initial_unis,
            'final_universities': final_unis,
            'initial_regions': initial_regions,
            'final_regions': final_regions,
            'universities_evolution': unis_evolution,
            'regions_evolution': regions_evolution,
            'mean_reward': np.mean(test_rewards),
            'final_reward': test_rewards[-1] if test_rewards else 0.0,
            'stability': np.std(test_rewards),
            'total_crisis_exposure': sum([c['total_intensity'] for c in test_crisis_impacts])
        }

        print(f"‚úÖ {alg_name} - –°—Ä–µ–¥–Ω—è—è –Ω–∞–≥—Ä–∞–¥–∞: {np.mean(test_rewards):.3f}, –°—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å: ¬±{np.std(test_rewards):.3f}")

    return final_results

# –≠–ö–°–ü–û–†–¢ –ò –í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø
def export_training_results(comparison_metrics, dataset, crisis_manager, env):
    print("üíæ –≠–ö–°–ü–û–†–¢ –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –û–ë–£–ß–ï–ù–ò–Ø...")

    # 1. –ò—Å—Ç–æ—Ä–∏—è –æ–±—É—á–µ–Ω–∏—è
    training_data = []
    for alg_name, metrics in comparison_metrics.items():
        for ep, reward in enumerate(metrics.get('episode_rewards', [])):
            training_data.append({
                'Algorithm': alg_name,
                'Episode': ep,
                'Reward': reward if np.isfinite(reward) else 0.0
            })

    pd.DataFrame(training_data).to_csv('results/data/training_100ep.csv', index=False)

    # 2. –°–≤–æ–¥–∫–∞ –∞–ª–≥–æ—Ä–∏—Ç–º–æ–≤
    summary_data = []
    for alg_name, metrics in comparison_metrics.items():
        rewards = [r for r in metrics.get('episode_rewards', []) if np.isfinite(r)]
        if rewards:
            summary_data.append({
                'Algorithm': alg_name,
                'Episodes': len(rewards),
                'Mean_Reward': np.mean(rewards),
                'Final_Reward': rewards[-1],
                'Max_Reward': np.max(rewards),
                'Min_Reward': np.min(rewards),
                'Std_Reward': np.std(rewards),
                'Training_Minutes': metrics.get('training_time', 0) / 60
            })

    pd.DataFrame(summary_data).to_csv('results/data/training_summary.csv', index=False)

    # 3. –ò—Å—Ç–æ—Ä–∏—è –∫—Ä–∏–∑–∏—Å–æ–≤ –ø–æ —à–∞–≥–∞–º
    if crisis_manager.step_timeline:
        crisis_timeline_data = []
        for entry in crisis_manager.step_timeline:
            crisis_timeline_data.append({
                'Step': entry['step'],
                'Crisis_Name': entry['crisis_name'],
                'Intensity': entry['intensity'],
                'Status': entry['status'],
                'Steps_Active': entry['steps_active'],
                'Duration': entry['duration']
            })

        pd.DataFrame(crisis_timeline_data).to_csv('results/data/crisis_timeline_steps.csv', index=False)

    # 4. –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –∫—Ä–∏–∑–∏—Å–æ–≤ –∏ –Ω–∞–≥—Ä–∞–¥
    if env.crisis_impact_history:
        correlation_data = []
        for entry in env.crisis_impact_history:
            correlation_data.append({
                'Episode': entry['episode'],
                'Total_Crisis_Intensity': entry['total_intensity'],
                'Num_Active_Crises': entry['num_active_crises'],
                'Active_Crises': ','.join(entry['active_crises']),
                'Mean_Reward': entry['mean_reward']
            })

        pd.DataFrame(correlation_data).to_csv('results/data/crisis_correlation.csv', index=False)

    print("‚úÖ –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –æ–±—É—á–µ–Ω–∏—è —ç–∫—Å–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω—ã")

def export_final_test_results(final_results):
    print("üíæ –≠–ö–°–ü–û–†–¢ –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –§–ò–ù–ê–õ–¨–ù–û–ì–û –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–Ø...")

    # 1. –ù–∞–≥—Ä–∞–¥—ã —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è
    final_rewards_data = []
    for alg_name, results in final_results.items():
        for step, reward in enumerate(results['test_rewards']):
            final_rewards_data.append({
                'Algorithm': alg_name,
                'Step': step,
                'Reward': reward,
                'Month': step + 1
            })

    pd.DataFrame(final_rewards_data).to_csv('results/final_test/final_test_rewards.csv', index=False)

    # 2. –ö—Ä–∏–∑–∏—Å—ã —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è
    final_crises_data = []
    for alg_name, results in final_results.items():
        for crisis_info in results['crisis_impacts']:
            final_crises_data.append({
                'Algorithm': alg_name,
                'Step': crisis_info['step'],
                'Total_Intensity': crisis_info['total_intensity'],
                'Num_Active_Crises': crisis_info['num_active_crises'],
                'Active_Crises': ','.join(crisis_info['active_crises'])
            })

    if final_crises_data:
        pd.DataFrame(final_crises_data).to_csv('results/final_test/final_test_crises.csv', index=False)

    # 3. –≠–≤–æ–ª—é—Ü–∏—è —É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç–æ–≤
    for alg_name, results in final_results.items():
        unis_data = []
        for snapshot in results['universities_evolution']:
            step = snapshot['step']
            unis = snapshot['universities']
            for uni_id in range(len(unis['performance'])):
                unis_data.append({
                    'Step': step,
                    'University_ID': uni_id,
                    'Performance': unis['performance'][uni_id],
                    'Financial': unis['financial'][uni_id],
                    'Market_Share': unis['market_share'][uni_id],
                    'Reputation': unis['reputation'][uni_id],
                    'Satisfaction': unis['satisfaction'][uni_id]
                })

        pd.DataFrame(unis_data).to_csv(f'results/final_test/universities_evolution_{alg_name}.csv', index=False)

    # 4. –≠–≤–æ–ª—é—Ü–∏—è —Ä–µ–≥–∏–æ–Ω–æ–≤
    for alg_name, results in final_results.items():
        regions_data = []
        for snapshot in results['regions_evolution']:
            step = snapshot['step']
            regions = snapshot['regions']
            for region_id in range(len(regions)):
                for program_id in range(len(regions[region_id])):
                    regions_data.append({
                        'Step': step,
                        'Region_ID': region_id,
                        'Program_ID': program_id,
                        'Demand': regions[region_id][program_id]
                    })

        pd.DataFrame(regions_data).to_csv(f'results/final_test/regions_evolution_{alg_name}.csv', index=False)

    # 5. –°–≤–æ–¥–∫–∞ —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è
    final_summary_data = []
    for alg_name, results in final_results.items():
        final_summary_data.append({
            'Algorithm': alg_name,
            'Mean_Reward': results['mean_reward'],
            'Final_Reward': results['final_reward'],
            'Stability': results['stability'],
            'Total_Crisis_Exposure': results['total_crisis_exposure'],
            'Performance_Improvement': (results['final_reward'] - results['test_rewards'][0]) if results['test_rewards'] else 0
        })

    pd.DataFrame(final_summary_data).to_csv('results/final_test/final_summary.csv', index=False)

    print("‚úÖ –†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è —ç–∫—Å–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω—ã")

def create_training_visualizations(comparison_metrics, crisis_manager, env):
    print("üìä –°–û–ó–î–ê–ù–ò–ï –ì–†–ê–§–ò–ö–û–í –û–ë–£–ß–ï–ù–ò–Ø...")

    plt.style.use('default')
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

    fig, axes = plt.subplots(2, 3, figsize=(18, 10))

    # 1. –ö—Ä–∏–≤—ã–µ –æ–±—É—á–µ–Ω–∏—è
    ax = axes[0, 0]
    for i, (alg_name, metrics) in enumerate(comparison_metrics.items()):
        rewards = [r if np.isfinite(r) else 0.0 for r in metrics.get('episode_rewards', [])]
        if rewards:
            episodes = list(range(len(rewards)))

            if len(rewards) > 10:
                window_size = 5
                smooth_rewards = pd.Series(rewards).rolling(window=window_size, center=True).mean()
                ax.plot(episodes, smooth_rewards, label=alg_name, color=colors[i], linewidth=3)

            ax.plot(episodes, rewards, color=colors[i], alpha=0.3, linewidth=1)

    ax.set_xlabel('–≠–ø–∏–∑–æ–¥')
    ax.set_ylabel('–ù–∞–≥—Ä–∞–¥–∞')
    ax.set_title('–ö—Ä–∏–≤—ã–µ –æ–±—É—á–µ–Ω–∏—è (100 —ç–ø–∏–∑–æ–¥–æ–≤)')
    ax.legend()
    ax.grid(True, alpha=0.3)

    # 2. –§–∏–Ω–∞–ª—å–Ω–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å
    ax = axes[0, 1]
    alg_names = []
    final_rewards = []
    std_rewards = []

    for alg_name, metrics in comparison_metrics.items():
        rewards = [r for r in metrics.get('episode_rewards', []) if np.isfinite(r)]
        if rewards:
            alg_names.append(alg_name)
            final_rewards.append(np.mean(rewards[-10:]) if len(rewards) >= 10 else np.mean(rewards))
            std_rewards.append(np.std(rewards[-10:]) if len(rewards) >= 10 else np.std(rewards))

    if alg_names:
        bars = ax.bar(alg_names, final_rewards, yerr=std_rewards, alpha=0.7,
                     color=colors[:len(alg_names)], capsize=5)
        ax.set_ylabel('–°—Ä–µ–¥–Ω—è—è –Ω–∞–≥—Ä–∞–¥–∞ (–ø–æ—Å–ª–µ–¥–Ω–∏–µ 10)')
        ax.set_title('–§–∏–Ω–∞–ª—å–Ω–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å')

        for bar, val, std_val in zip(bars, final_rewards, std_rewards):
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + std_val + 0.05,
                   f'{val:.2f}¬±{std_val:.2f}', ha='center', va='bottom', fontsize=9)

    # 3. –°—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å
    ax = axes[0, 2]
    stability_scores = []
    for metrics in comparison_metrics.values():
        rewards = [r for r in metrics.get('episode_rewards', []) if np.isfinite(r)]
        stability_scores.append(np.std(rewards) if rewards else 0.0)

    if alg_names:
        bars = ax.bar(alg_names, stability_scores, alpha=0.7, color=colors[:len(alg_names)])
        ax.set_ylabel('–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ')
        ax.set_title('–°—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å (–º–µ–Ω—å—à–µ = –ª—É—á—à–µ)')

        for bar, val in zip(bars, stability_scores):
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                   f'{val:.3f}', ha='center', va='bottom')

    # 4. –•–∞–æ—Ç–∏—á–µ—Å–∫–∏–µ –∫—Ä–∏–∑–∏—Å—ã –ø–æ —à–∞–≥–∞–º (–ø–µ—Ä–≤—ã–µ 120 —à–∞–≥–æ–≤)
    ax = axes[1, 0]
    crisis_steps = []
    crisis_intensities = []

    for step, crisis in crisis_manager.step_based_crises.items():
        if step < 120:  # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –ø–µ—Ä–≤—ã–µ 120 —à–∞–≥–æ–≤
            crisis_steps.append(step)
            crisis_intensities.append(crisis['intensity'])

    if crisis_steps:
        ax.scatter(crisis_steps, crisis_intensities, alpha=0.7, color='red', s=50)
        ax.set_xlabel('–®–∞–≥ (–º–µ—Å—è—Ü)')
        ax.set_ylabel('–ò–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å –∫—Ä–∏–∑–∏—Å–∞')
        ax.set_title('–•–∞–æ—Ç–∏—á–µ—Å–∫–∏–µ –∫—Ä–∏–∑–∏—Å—ã (120 –º–µ—Å—è—Ü–µ–≤)')
        ax.set_xlim(0, 120)
        ax.set_ylim(0, 1.0)
        ax.grid(True, alpha=0.3)

    # 5. –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –∫—Ä–∏–∑–∏—Å–æ–≤ –∏ –Ω–∞–≥—Ä–∞–¥
    ax = axes[1, 1]
    if env.crisis_impact_history:
        crisis_intensities = [entry['total_intensity'] for entry in env.crisis_impact_history]
        mean_rewards = [entry['mean_reward'] for entry in env.crisis_impact_history]

        if crisis_intensities and mean_rewards:
            ax.scatter(crisis_intensities, mean_rewards, alpha=0.7, s=60)
            ax.set_xlabel('–ò–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å –∫—Ä–∏–∑–∏—Å–æ–≤')
            ax.set_ylabel('–°—Ä–µ–¥–Ω—è—è –Ω–∞–≥—Ä–∞–¥–∞')
            ax.set_title('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è: –ö—Ä–∏–∑–∏—Å—ã vs –ù–∞–≥—Ä–∞–¥—ã')
            ax.grid(True, alpha=0.3)

            if len(crisis_intensities) > 2:
                z = np.polyfit(crisis_intensities, mean_rewards, 1)
                p = np.poly1d(z)
                x_trend = np.linspace(min(crisis_intensities), max(crisis_intensities), 100)
                ax.plot(x_trend, p(x_trend), "r--", alpha=0.8)

    # 6. –í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è
    ax = axes[1, 2]
    training_times = [metrics.get('training_time', 0)/60 for metrics in comparison_metrics.values()]
    if alg_names:
        bars = ax.bar(alg_names, training_times, alpha=0.7, color=colors[:len(alg_names)])
        ax.set_ylabel('–í—Ä–µ–º—è (–º–∏–Ω—É—Ç—ã)')
        ax.set_title('–í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è')

        for bar, val in zip(bars, training_times):
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
                   f'{val:.0f}–º', ha='center', va='bottom')

    plt.tight_layout()
    plt.savefig('results/plots/training_results.png', dpi=300, bbox_inches='tight')
    plt.show()

    print("‚úÖ –ì—Ä–∞—Ñ–∏–∫–∏ –æ–±—É—á–µ–Ω–∏—è —Å–æ–∑–¥–∞–Ω—ã")

def create_final_test_visualizations(final_results):
    print("üìä –°–û–ó–î–ê–ù–ò–ï –ì–†–ê–§–ò–ö–û–í –§–ò–ù–ê–õ–¨–ù–û–ì–û –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–Ø...")

    plt.style.use('default')
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

    # –û—Å–Ω–æ–≤–Ω–æ–π –≥—Ä–∞—Ñ–∏–∫ —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è (2x3)
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))

    # 1. –ù–∞–≥—Ä–∞–¥—ã —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –ø–æ —à–∞–≥–∞–º
    ax = axes[0, 0]
    for i, (alg_name, results) in enumerate(final_results.items()):
        steps = list(range(len(results['test_rewards'])))
        rewards = results['test_rewards']

        # –°–≥–ª–∞–∂–∏–≤–∞–Ω–∏–µ
        if len(rewards) > 10:
            window_size = 5
            smooth_rewards = pd.Series(rewards).rolling(window=window_size, center=True).mean()
            ax.plot(steps, smooth_rewards, label=alg_name, color=colors[i], linewidth=3)

        ax.plot(steps, rewards, color=colors[i], alpha=0.3, linewidth=1)

    ax.set_xlabel('–®–∞–≥ (–º–µ—Å—è—Ü)')
    ax.set_ylabel('–ù–∞–≥—Ä–∞–¥–∞')
    ax.set_title('–§–∏–Ω–∞–ª—å–Ω–æ–µ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ (120 –º–µ—Å—è—Ü–µ–≤)')
    ax.legend()
    ax.grid(True, alpha=0.3)

    # 2. –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ñ–∏–Ω–∞–ª—å–Ω–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏
    ax = axes[0, 1]
    alg_names = list(final_results.keys())
    mean_rewards = [results['mean_reward'] for results in final_results.values()]
    stabilities = [results['stability'] for results in final_results.values()]

    bars = ax.bar(alg_names, mean_rewards, yerr=stabilities, alpha=0.7,
                 color=colors[:len(alg_names)], capsize=5)
    ax.set_ylabel('–°—Ä–µ–¥–Ω—è—è –Ω–∞–≥—Ä–∞–¥–∞')
    ax.set_title('–ü—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –≤ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–∏')

    for bar, val, std_val in zip(bars, mean_rewards, stabilities):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + std_val + 0.05,
               f'{val:.2f}¬±{std_val:.2f}', ha='center', va='bottom', fontsize=9)

    # 3. –ö—Ä–∏–∑–∏—Å—ã –≤–æ –≤—Ä–µ–º—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è
    ax = axes[0, 2]

    # –°–æ–±–∏—Ä–∞–µ–º –≤—Å–µ –∫—Ä–∏–∑–∏—Å—ã –∏–∑ –ø–µ—Ä–≤–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ (–æ–Ω–∏ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –¥–ª—è –≤—Å–µ—Ö)
    first_alg = list(final_results.keys())[0]
    crisis_steps = []
    crisis_intensities = []

    for crisis_info in final_results[first_alg]['crisis_impacts']:
        if crisis_info['total_intensity'] > 0:
            crisis_steps.append(crisis_info['step'])
            crisis_intensities.append(crisis_info['total_intensity'])

    if crisis_steps:
        ax.scatter(crisis_steps, crisis_intensities, alpha=0.7, color='red', s=40)
        ax.set_xlabel('–®–∞–≥ (–º–µ—Å—è—Ü)')
        ax.set_ylabel('–ò–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å –∫—Ä–∏–∑–∏—Å–∞')
        ax.set_title('–ö—Ä–∏–∑–∏—Å—ã –≤ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–∏')
        ax.set_xlim(0, 120)
        ax.grid(True, alpha=0.3)

    # 4. –≠–≤–æ–ª—é—Ü–∏—è —Å—Ä–µ–¥–Ω–µ–π –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç–æ–≤
    ax = axes[1, 0]
    for i, (alg_name, results) in enumerate(final_results.items()):
        steps = []
        mean_performance = []

        for snapshot in results['universities_evolution']:
            steps.append(snapshot['step'])
            mean_performance.append(np.mean(snapshot['universities']['performance']))

        if steps:
            ax.plot(steps, mean_performance, label=alg_name, color=colors[i], linewidth=2, marker='o')

    ax.set_xlabel('–®–∞–≥ (–º–µ—Å—è—Ü)')
    ax.set_ylabel('–°—Ä–µ–¥–Ω—è—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –≤—É–∑–æ–≤')
    ax.set_title('–≠–≤–æ–ª—é—Ü–∏—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç–æ–≤')
    ax.legend()
    ax.grid(True, alpha=0.3)

    # 5. –≠–≤–æ–ª—é—Ü–∏—è —Ñ–∏–Ω–∞–Ω—Å–æ–≤–æ–≥–æ —Å–æ—Å—Ç–æ—è–Ω–∏—è —É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç–æ–≤
    ax = axes[1, 1]
    for i, (alg_name, results) in enumerate(final_results.items()):
        steps = []
        mean_financial = []

        for snapshot in results['universities_evolution']:
            steps.append(snapshot['step'])
            mean_financial.append(np.mean(snapshot['universities']['financial']))

        if steps:
            ax.plot(steps, mean_financial, label=alg_name, color=colors[i], linewidth=2, marker='s')

    ax.set_xlabel('–®–∞–≥ (–º–µ—Å—è—Ü)')
    ax.set_ylabel('–°—Ä–µ–¥–Ω–µ–µ —Ñ–∏–Ω–∞–Ω—Å–æ–≤–æ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ')
    ax.set_title('–≠–≤–æ–ª—é—Ü–∏—è —Ñ–∏–Ω–∞–Ω—Å–æ–≤–æ–≥–æ —Å–æ—Å—Ç–æ—è–Ω–∏—è')
    ax.legend()
    ax.grid(True, alpha=0.3)

    # 6. –≠–≤–æ–ª—é—Ü–∏—è —Ä–µ–≥–∏–æ–Ω–∞–ª—å–Ω—ã—Ö —Å–æ—Å—Ç–æ—è–Ω–∏–π
    ax = axes[1, 2]
    for i, (alg_name, results) in enumerate(final_results.items()):
        steps = []
        mean_regional = []

        for snapshot in results['regions_evolution']:
            steps.append(snapshot['step'])
            mean_regional.append(np.mean(snapshot['regions']))

        if steps:
            ax.plot(steps, mean_regional, label=alg_name, color=colors[i], linewidth=2, marker='^')

    ax.set_xlabel('–®–∞–≥ (–º–µ—Å—è—Ü)')
    ax.set_ylabel('–°—Ä–µ–¥–Ω–∏–π —Ä–µ–≥–∏–æ–Ω–∞–ª—å–Ω—ã–π —Å–ø—Ä–æ—Å')
    ax.set_title('–≠–≤–æ–ª—é—Ü–∏—è —Ä–µ–≥–∏–æ–Ω–∞–ª—å–Ω—ã—Ö —Ä—ã–Ω–∫–æ–≤')
    ax.legend()
    ax.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('results/plots/final_test_results.png', dpi=300, bbox_inches='tight')
    plt.show()

    # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫: –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –Ω–∞—á–∞–ª—å–Ω—ã—Ö –∏ —Ñ–∏–Ω–∞–ª—å–Ω—ã—Ö —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))

    metrics = ['performance', 'financial', 'market_share']
    metric_names = ['–ü—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å', '–§–∏–Ω–∞–Ω—Å—ã', '–î–æ–ª—è —Ä—ã–Ω–∫–∞']

    for idx, (metric, metric_name) in enumerate(zip(metrics, metric_names)):
        ax = axes[idx]

        x_pos = np.arange(len(final_results))
        width = 0.35

        initial_values = []
        final_values = []

        for alg_name, results in final_results.items():
            if metric == 'performance':
                initial_val = np.mean(np.mean(results['initial_universities']['performance_indicators'], axis=1))
                final_val = np.mean(np.mean(results['final_universities']['performance_indicators'], axis=1))
            else:
                initial_val = np.mean(results['initial_universities'][metric.replace('performance', 'performance_indicators')])
                final_val = np.mean(results['final_universities'][metric.replace('performance', 'performance_indicators')])

            initial_values.append(initial_val)
            final_values.append(final_val)

        bars1 = ax.bar(x_pos - width/2, initial_values, width, label='–ù–∞—á–∞–ª—å–Ω–æ–µ', alpha=0.7)
        bars2 = ax.bar(x_pos + width/2, final_values, width, label='–§–∏–Ω–∞–ª—å–Ω–æ–µ', alpha=0.7)

        ax.set_xlabel('–ê–ª–≥–æ—Ä–∏—Ç–º')
        ax.set_ylabel(metric_name)
        ax.set_title(f'–ò–∑–º–µ–Ω–µ–Ω–∏–µ: {metric_name}')
        ax.set_xticks(x_pos)
        ax.set_xticklabels(list(final_results.keys()))
        ax.legend()

        # –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã
        for bar, val in zip(bars1, initial_values):
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                   f'{val:.2f}', ha='center', va='bottom', fontsize=8)

        for bar, val in zip(bars2, final_values):
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                   f'{val:.2f}', ha='center', va='bottom', fontsize=8)

    plt.tight_layout()
    plt.savefig('results/plots/final_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()

    print("‚úÖ –ì—Ä–∞—Ñ–∏–∫–∏ —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è —Å–æ–∑–¥–∞–Ω—ã")

# –û–°–ù–û–í–ù–û–ô –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢
def run_experiment():
    print("üöÄ –ó–ê–ü–£–°–ö –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢–ê")

    comparison_metrics = {}
    algorithms = {
        'Baseline': MADDPG_Baseline,
        'EVO': MADDPG_EVO,
        'EVO_DGM': MADDPG_EVO_DGM
    }

    trained_models = {}

    for alg_name, AlgClass in algorithms.items():
        print(f"\\n{'='*60}")
        print(f"üéØ –û–ë–£–ß–ï–ù–ò–ï: {alg_name}")
        print('='*60)

        algorithm = AlgClass(config.N_UNIVERSITIES, config.STATE_DIM, config.ACTION_DIM)
        algorithm_metrics = {'episode_rewards': [], 'training_time': 0}
        performance_history = []

        start_time = datetime.now()

        for episode in tqdm(range(config.MAX_EPISODES), desc=f"–û–±—É—á–µ–Ω–∏–µ {alg_name}"):
            states = env.reset()

            # –≠–≤–æ–ª—é—Ü–∏—è (—Ç–æ–ª—å–∫–æ –¥–ª—è EVO –∏ EVO_DGM)
            if hasattr(algorithm, 'evolve_population'):
                current_performance = algorithm_metrics['episode_rewards'][-5:] if len(algorithm_metrics['episode_rewards']) >= 5 else [0.0] * 5
                algorithm.evolve_population(episode, current_performance)

                # –û–±–Ω–æ–≤–ª–µ–Ω–∏–µ DGM –º–∞—Ç—Ä–∏—Ü—ã (—Ç–æ–ª—å–∫–æ –¥–ª—è EVO_DGM)
                if hasattr(algorithm, '_update_communication_matrix'):
                    algorithm._update_communication_matrix(performance_history)

            step_rewards = []

            for step in range(config.MAX_STEPS):
                actions = algorithm.select_actions(states)
                next_states, rewards, done, info = env.step(actions)

                algorithm.step(states, actions, rewards, next_states, done)

                step_reward = np.mean([r if np.isfinite(r) else 0.0 for r in rewards])
                step_rewards.append(step_reward)

                if len(algorithm.replay_buffer) > config.BATCH_SIZE:
                    algorithm.update()

                states = next_states
                if done:
                    break

            episode_reward = np.mean([r for r in step_rewards if np.isfinite(r)]) if step_rewards else 0.0
            algorithm_metrics['episode_rewards'].append(episode_reward)
            performance_history.append([episode_reward] * config.N_UNIVERSITIES)

            if episode % 25 == 0:
                avg_reward = np.mean(algorithm_metrics['episode_rewards'][-5:])

                evolution_info = ""
                if hasattr(algorithm, 'evolution_triggered') and algorithm.evolution_triggered:
                    recent_evolutions = [e for e in algorithm.evolution_triggered if e['episode'] >= episode-25]
                    if recent_evolutions:
                        evolution_info = f" | –≠–≤–æ–ª—é—Ü–∏–π: {len(recent_evolutions)}"

                print(f"–≠–ø.{episode:2d}: –ù–∞–≥—Ä–∞–¥–∞={episode_reward:5.2f} | –°—Ä–µ–¥.={avg_reward:5.2f}{evolution_info}")

        end_time = datetime.now()
        algorithm_metrics['training_time'] = (end_time - start_time).total_seconds()

        rewards = algorithm_metrics['episode_rewards']
        print(f"\\n‚úÖ {alg_name} –∑–∞–≤–µ—Ä—à–µ–Ω –∑–∞ {algorithm_metrics['training_time']/60:.1f} –º–∏–Ω—É—Ç")
        print(f"üìä –°—Ä–µ–¥–Ω—è—è: {np.mean(rewards):.3f} | –§–∏–Ω–∞–ª—å–Ω–∞—è: {rewards[-1]:.3f} | " +
              f"–ú–∞–∫—Å: {np.max(rewards):.3f} | –°—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å: ¬±{np.std(rewards):.3f}")

        # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
        if hasattr(algorithm, 'evolution_triggered'):
            print(f"üß¨ –≠–≤–æ–ª—é—Ü–∏–π: {len(algorithm.evolution_triggered)}")
            if algorithm.evolution_triggered:
                reasons = [e['reason'] for e in algorithm.evolution_triggered]
                reason_counts = {r: reasons.count(r) for r in set(reasons)}
                print(f"   –ü—Ä–∏—á–∏–Ω—ã: {reason_counts}")

        if hasattr(algorithm, 'communication_history'):
            print(f"üì° –ö–æ–º–º—É–Ω–∏–∫–∞—Ü–∏–æ–Ω–Ω—ã—Ö –æ–±–º–µ–Ω–æ–≤: {len(algorithm.communication_history)}")

        comparison_metrics[alg_name] = algorithm_metrics
        algorithm.save_models(f"results/models/{alg_name}_step_based")
        trained_models[alg_name] = algorithm

    return comparison_metrics, trained_models

# –ì–õ–ê–í–ù–ê–Ø –§–£–ù–ö–¶–ò–Ø
def main():
    print("üéØ MADDPG –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢ –° –•–ê–û–¢–ò–ß–ï–°–ö–ò–ú–ò –ö–†–ò–ó–ò–°–ê–ú–ò")
    print(f"üìÖ {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    print(f"üé≤ SEED: {SEED}, üì± –£—Å—Ç—Ä–æ–π—Å—Ç–≤–æ: {DEVICE}")
    print("="*80)

    try:
        # –≠–¢–ê–ü 1: –û–ë–£–ß–ï–ù–ò–ï
        print("üöÄ –≠–¢–ê–ü 1: –û–ë–£–ß–ï–ù–ò–ï –ù–ê 100 –≠–ü–ò–ó–û–î–ê–•")
        comparison_metrics, trained_models = run_experiment()

        # –≠–¢–ê–ü 2: –≠–ö–°–ü–û–†–¢ –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –û–ë–£–ß–ï–ù–ò–Ø
        print("\\nüíæ –≠–¢–ê–ü 2: –≠–ö–°–ü–û–†–¢ –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –û–ë–£–ß–ï–ù–ò–Ø")
        export_training_results(comparison_metrics, dataset, crisis_manager, env)

        # –≠–¢–ê–ü 3: –ì–†–ê–§–ò–ö–ò –û–ë–£–ß–ï–ù–ò–Ø
        print("\\nüìä –≠–¢–ê–ü 3: –í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –û–ë–£–ß–ï–ù–ò–Ø")
        create_training_visualizations(comparison_metrics, crisis_manager, env)

        # –≠–¢–ê–ü 4: –§–ò–ù–ê–õ–¨–ù–û–ï –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–ï
        print("\\nüéØ –≠–¢–ê–ü 4: –§–ò–ù–ê–õ–¨–ù–û–ï –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–ï –û–ë–£–ß–ï–ù–ù–´–• –ú–û–î–ï–õ–ï–ô")
        final_results = run_final_test(trained_models, dataset, crisis_manager, env)

        # –≠–¢–ê–ü 5: –≠–ö–°–ü–û–†–¢ –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–Ø
        print("\\nüíæ –≠–¢–ê–ü 5: –≠–ö–°–ü–û–†–¢ –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–Ø")
        export_final_test_results(final_results)

        # –≠–¢–ê–ü 6: –ì–†–ê–§–ò–ö–ò –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–Ø
        print("\\nüìä –≠–¢–ê–ü 6: –í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–Ø")
        create_final_test_visualizations(final_results)

        # –ò–¢–û–ì–û–í–´–ô –û–¢–ß–ï–¢
        print("\\nüéØ –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢ –ó–ê–í–ï–†–®–ï–ù!")
        print("="*80)

        total_episodes = sum(len(metrics['episode_rewards']) for metrics in comparison_metrics.values())
        total_time = sum(metrics['training_time'] for metrics in comparison_metrics.values()) / 60

        print(f"üìä –ê–ª–≥–æ—Ä–∏—Ç–º–æ–≤: {len(comparison_metrics)}")
        print(f"üéØ –≠–ø–∏–∑–æ–¥–æ–≤ –æ–±—É—á–µ–Ω–∏—è: {total_episodes}")
        print(f"‚è±Ô∏è  –í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è: {total_time:.1f} –º–∏–Ω—É—Ç")
        print(f"üí• –•–∞–æ—Ç–∏—á–µ—Å–∫–∏—Ö –∫—Ä–∏–∑–∏—Å–æ–≤: {len(crisis_manager.step_based_crises)}")
        print(f"üìà –®–∞–≥–æ–≤ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è: {config.MAX_STEPS}")

        print("\\nüìÅ –°–û–ó–î–ê–ù–û:")
        print("   üìä –§–∞–π–ª—ã –æ–±—É—á–µ–Ω–∏—è –≤ results/data/")
        print("   üìä –§–∞–π–ª—ã —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –≤ results/final_test/")
        print("   üìà –ì—Ä–∞—Ñ–∏–∫–∏ –æ–±—É—á–µ–Ω–∏—è –≤ results/plots/training_results.png")
        print("   üìà –ì—Ä–∞—Ñ–∏–∫–∏ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –≤ results/plots/final_test_results.png")
        print("   ü§ñ –û–±—É—á–µ–Ω–Ω—ã–µ –º–æ–¥–µ–ª–∏ –≤ results/models/")

        print("\\nüèÜ –†–ï–ó–£–õ–¨–¢–ê–¢–´ –û–ë–£–ß–ï–ù–ò–Ø:")
        for alg_name, metrics in comparison_metrics.items():
            rewards = metrics['episode_rewards']
            improvement = rewards[-1] - rewards[0] if len(rewards) > 1 else 0
            print(f"  {alg_name:<12}: {np.mean(rewards):.3f} (—Å—Ä–µ–¥–Ω—è—è), {rewards[-1]:.3f} (—Ñ–∏–Ω–∞–ª—å–Ω–∞—è), " +
                  f"¬±{np.std(rewards):.3f} (—Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å), +{improvement:.3f} (–ø—Ä–æ–≥—Ä–µ—Å—Å)")

        print("\\nüéØ –†–ï–ó–£–õ–¨–¢–ê–¢–´ –§–ò–ù–ê–õ–¨–ù–û–ì–û –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–Ø:")
        for alg_name, results in final_results.items():
            print(f"  {alg_name:<12}: {results['mean_reward']:.3f} (—Å—Ä–µ–¥–Ω—è—è), {results['final_reward']:.3f} (—Ñ–∏–Ω–∞–ª—å–Ω–∞—è), " +
                  f"¬±{results['stability']:.3f} (—Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å), {results['total_crisis_exposure']:.1f} (–∫—Ä–∏–∑–∏—Å—ã)")

        return comparison_metrics, final_results

    except Exception as e:
        print(f"‚ùå –û–®–ò–ë–ö–ê: {e}")
        import traceback
        traceback.print_exc()
        return None, None

    finally:
        print("\\nüéØ –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢ –ó–ê–í–ï–†–®–ï–ù!")

if __name__ == "__main__":
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–æ–≤
    dataset_generator = DatasetGenerator(SEED)
    dataset = dataset_generator.generate_full_dataset()

    crisis_manager = ChaoticCrisisManager(SEED)
    env = UniversityEnvironment(dataset, crisis_manager)

    # –ó–∞–ø—É—Å–∫ –ø–æ–ª–Ω–æ–≥–æ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞
    training_results, test_results = main()
```
